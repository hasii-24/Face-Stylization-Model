```python
def generate_style(input_img, style_choice, seed, style_strength):
    try:
        print("▶ Called with:", input_img, style_choice)

        torch.manual_seed(seed)

        # 1️⃣ Align face (expects filepath)
        aligned_face = align_face(input_img)  # PIL Image

        # 2️⃣ e4e projection
        my_w = e4e_projection(aligned_face, "input.pt", device).unsqueeze(0)

        # 3️⃣ Load style checkpoint
        style_ckpt = torch.load(f"models/{style_choice}.pt", map_location=device)

        # 4️⃣ Detect resolution
        args = style_ckpt.get("args", None)
        size = args.size if args else 1024

        # 5️⃣ Load generator
        generator = Generator(size, 512, 8, 2).to(device)
        generator.load_state_dict(style_ckpt["g"], strict=False)
        generator.eval()

        # 6️⃣ Generate image
        with torch.no_grad():
            styled = generator(my_w, input_is_latent=True)

        # 7️⃣ Resize helpers
        resize = transforms.Compose([
            transforms.Resize((size, size)),
            transforms.ToTensor(),
            transforms.Normalize((0.5,)*3, (0.5,)*3),
        ])

        face = resize(aligned_face).unsqueeze(0).to(device)

        # 8️⃣ Style reference
        style_path = f"style_images_aligned/{style_choice}.png"
        if not os.path.exists(style_path):
            raise FileNotFoundError(style_path)

        style_img = resize(Image.open(style_path)).unsqueeze(0).to(device)

        # 9️⃣ Blend (strong style)
        output = face * (1 - style_strength) + styled * style_strength

        # 🔟 Grid
        grid = utils.make_grid(
            torch.cat([style_img, face, output], 0),
            normalize=True,
            value_range=(-1, 1)
        )

        grid = (grid.permute(1, 2, 0).cpu().numpy() * 255).astype("uint8")
        return Image.fromarray(grid)

    except Exception as e:
        print("❌ ERROR:", e)
        return Image.new("RGB", (512, 512), "red")
```

```python
import os
os.chdir("/content/JoJoGAN")

import torch
from copy import deepcopy
from torchvision import utils, transforms
from PIL import Image
import gradio as gr

from util import align_face, strip_path_extension
from model import Generator
from e4e_projection import projection

# --------------------------------------------------
# Device
# --------------------------------------------------
device = "cuda" if torch.cuda.is_available() else "cpu"

# --------------------------------------------------
# Transform
# --------------------------------------------------
transform = transforms.Compose([
    transforms.Resize((1024, 1024)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))
])

# --------------------------------------------------
# Styles
# --------------------------------------------------
style_choices = ["jojo","disney","arcane","arcane_jinx","art"]

# --------------------------------------------------
# Load base generator ONCE
# --------------------------------------------------
latent_dim = 512
base_generator = Generator(1024, latent_dim, 8, 2).to(device)
ckpt = torch.load("models/stylegan2-ffhq-config-f.pt", map_location=device)
base_generator.load_state_dict(ckpt["g_ema"], strict=False)
base_generator.eval()

# --------------------------------------------------
# Load styled generator
# --------------------------------------------------
def load_style_generator(style):
    gen = deepcopy(base_generator)
    ckpt = torch.load(f"models/{style}.pt", map_location=device)
    gen.load_state_dict(ckpt["g"], strict=False)
    gen.eval().to(device)
    return gen

# --------------------------------------------------
# Helper: tensor grid → PIL
# --------------------------------------------------
def tensor_to_pil(grid):
    grid = (grid.permute(1,2,0).cpu().numpy() * 255).astype("uint8")
    return Image.fromarray(grid)

# ==================================================
# SINGLE STYLE → 3 images
# ==================================================
def single_style_fn(input_img, style, seed):
    torch.manual_seed(seed)

    aligned_face = align_face(input_img)
    w = projection(aligned_face, "single.pt", device).unsqueeze(0)

    generator = load_style_generator(style)

    with torch.no_grad():
        out = generator(w, input_is_latent=True)

    style_img = transform(
        Image.open(f"style_images_aligned/{style}.png")
    ).unsqueeze(0).to(device)

    face = transform(aligned_face).unsqueeze(0).to(device)

    # [Style | Input | Output]
    grid = torch.cat([style_img, face, out], 0)
    grid = utils.make_grid(grid, nrow=3, normalize=True, value_range=(-1,1))

    return tensor_to_pil(grid)

# ==================================================
# STYLE BLENDING → 4 images
# ==================================================
def blend_style_fn(input_img, styles, alpha1, alpha2):
    if len(styles) != 2:
        raise gr.Error("Please select exactly 2 styles")

    style1, style2 = styles

    aligned_face = align_face(input_img)
    w = projection(aligned_face, "blend.pt", device).unsqueeze(0)

    gen1 = load_style_generator(style1)
    gen2 = load_style_generator(style2)

    with torch.no_grad():
        out1 = gen1(w, input_is_latent=True)
        out2 = gen2(w, input_is_latent=True)

    blended = alpha1 * out1 + alpha2 * out2
    blended = torch.clamp(blended, -1, 1)

    style_img1 = transform(
        Image.open(f"style_images_aligned/{style1}.png")
    ).unsqueeze(0).to(device)

    style_img2 = transform(
        Image.open(f"style_images_aligned/{style2}.png")
    ).unsqueeze(0).to(device)

    face = transform(aligned_face).unsqueeze(0).to(device)

    # [StyleA | StyleB | Input | Blended]
    grid = torch.cat([style_img1, style_img2, face, blended], 0)
    grid = utils.make_grid(grid, nrow=4, normalize=True, value_range=(-1,1))

    return tensor_to_pil(grid)

# ==================================================
# GRADIO UI
# ==================================================
with gr.Blocks(title="ONE SHOT FACE STYLIZATION") as demo:

    gr.Markdown("# 🎭 JoJoGAN Style Studio")
    gr.Markdown("Single Style (3 imgs) & Style Blending (4 imgs)")

    with gr.Row():

        # ---------------- SINGLE STYLE ----------------
        with gr.Column():
            gr.Markdown("## 🎨 Single Style Transfer")

            single_img = gr.Image(type="filepath", label="Upload Face")
            single_style = gr.Dropdown(style_choices, value="disney", label="Style")
            single_seed = gr.Slider(0, 9999, value=2000, label="Seed")

            single_btn = gr.Button("Generate Single Style")
            single_out = gr.Image(label="Style | Input | Output")

            single_btn.click(
                single_style_fn,
                inputs=[single_img, single_style, single_seed],
                outputs=single_out
            )

        # ---------------- STYLE BLENDING ----------------
        with gr.Column():
            gr.Markdown("## 🔥 Style Blending")

            blend_img = gr.Image(type="filepath", label="Upload Face")
            blend_styles = gr.CheckboxGroup(
                choices=style_choices,
                label="Select 2 Styles",
            )
            alpha1 = gr.Slider(0, 1, value=0.5, step=0.05, label="Alpha Style A")
            alpha2 = gr.Slider(0, 1, value=0.5, step=0.05, label="Alpha Style B")

            blend_btn = gr.Button("Generate Blend")
            blend_out = gr.Image(label="StyleA | StyleB | Input | Blended")

            blend_btn.click(
                blend_style_fn,
                inputs=[blend_img, blend_styles, alpha1, alpha2],
                outputs=blend_out
            )

demo.launch(share=True, debug=True, show_error=True)
```
